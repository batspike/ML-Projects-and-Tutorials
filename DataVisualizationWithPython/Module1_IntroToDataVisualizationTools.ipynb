# Overview of Data Visualization

### What is Data Visualization ?
* Graphical Representation of data and information.
* Can take a variety of forms from simple line to pie charts.
* Basic Charts and Graphs are the simplest form in representing numerical data.
* Interactive Dashboards and Maps are more complex in representation of real time information.

### Why build visuals?
* help us to easily understand data.
    * highlights patterns, trends, and relationships that may not be obvious at numerical state.
* Communicate Insights
    * Simplifies the data for stakeholders in storytelling.

### Identify trends and data patterns
* helps gain new insights.

### Present data for better understanding and interpretation
* letting us make more informed choices.

### When creating a visual, always remember:
* Less is more effective
* Less is more attractive
* Less is more impactfull

### Types of Plots
#### Different types of plots:
* Line plot
    * Displays data as a series of data points connected by straight lines.
    * Display trends over time.
    * Compare data sets with Continuous Independent Variable.
    * Use Cases:
        * Illustrate cause-and-effect relationships.
        * Visualize continuous data.
    * The scale of the dependent axis can emphasize the intensity of the message.
* Bar plot
    * Displays data using rectangular bars.
    * The height or length of the bars represents the magnitude of the data.
    * Display Bars can be oriented Vertically or Horizontally.
    * Use Cases:
        * Compare different Categories or Groups.
        * Display Discrete data that has distinct categories.
        * Show how different categories contribute to a whole.
        * Visualize data that can easily be Ranked or Ordered.
    * Ensure dependent Y-axis starts at zero for realistic representation.
* Scatter plot
    * Use cartesian coordinates to display values for two variables.
    * Value of one variable determines the position on the horizontal axis, and value of another variable for vertical axis.
    * Use Cases:
        * Use in Examining the relationship between two continuous variables.
        * For investigating patterns or treands in data.
        * Detect outliers or unusual observations.
        * Identify clusters or groups in the data.
        * Exploring complex data.
* Box plot
    * Displays the distribution of a dataset along with key statistical measures.
    * Represents the Interquartile range as a box, a Median line, and Whiskers indicating range.
    * Represents Outliers as individual data points beyond whiskers.
    * Use Cases:
        * Compare the distribution of a continuous variable across different Categories or Groups.e.g Comparing Salaries across Departments.
        * Examine Spread and Skewness of a dataset, visualizing quartiles.
        * Identify and analyze potential outliers.
        * Visualize summary statistics.
        * Comapre distributions of multiple variables in datasets.
* Histogram
    * Graphical representation of dataset distributions.
    * Shows Frequency or Relative Frequency within intervals.
    * Bars represent the data count in eact intervale.
    * Use Cases:
        * Understand the overall data distribution.
        * Visually depict the shape of the data.
        * Assess skewness in the data.
        * Showcase data Variability and Spread.
    * Ensure optimal binnings in data for clearer visual. Too few or too many bins can over simplify or over complicate.

### Introduction to Matplotlib
#### Matplotlib Architecture
* Backend Layer
    * FigureCanvas encompass the area onto which the figure is drawn.
    * Renderer knows how to draw on the FigureCanvas.
    * Event handles user inputs from keyboard and mouse clicks.

* Artist Layer
    * Artist knows how to use the Renderer to draw on the canvas.
    
* Scripting Layer
    * pyplot



```python
# generate a histogram of  100 random values using Pyplot interface.

import matplotlib.pyplot as plt # import the scripting interface
import numpy as np
plt.style.use( {'dark_background','seaborn-v0_8-whitegrid'})

x = np.random.randn(10000)
plt.hist(x, 100)
plt.title(r'Normal Distribution with $\mu = 0, \sigma=1$')
plt.savefig('matplotlib_histogram.png')
plt.show()
```

### Basic Plotting with Matplotlib
* once the plot is rendered via plt.show(), we cannot amend the plot.
* a new plot is needed after the plot is generated, thus any changes/plots/labels we need must be specified before rendering the plot with show() method.

```python
plt.plot(5,5,'o') # put a point at (5,5)
plt.show()
```

```python
# a new plot is needed after the plot is generated, thus any changes/plots/labels we need must be specified before rendering the plot with show() method.
plt.plot(5,5,'o') # put a point at (5,5)
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Plotting Example")

plt.show()
```

### Pandas library also has an implementation of the Matplotlib library
* In this case, the plot object must be created first, and then we applied any changes to it.

```python
# Using the Pandas implementation of the Matplotlib plot() method.

import pandas as pd

pokemons = pd.read_csv("data/pokemon.csv")
types_count = pokemons["Type"].value_counts().head(10)

types_count.plot(kind='bar', title="Pokemon Types Count").set(xlabel="Pokemon Type", ylabel="Count")
```
