## Basic and Specialized Visualization Tools

### Basic Visualization Tools
* Area Plots
    * Also known as area chart or graph
    * Displays magnitude and proportion of multiple variables over a continuous axis, typically x-axis
    * Represents time or another ordered dimension
    * Is based on the line plot
    * Use Cases:
        * tracking stock market performance.
        * visualizing population demographics.
        * displaying the distribution of resources.
* Histograms
    * Represents the frequency distribution of a dataset.
        * partitions numeric data into bins.
        * assigns each data point in the dataset to a bin.
        * counts the number of data points per bin.
* Bar Charts
    * Also known as bar graph, the size of the bar is proportional to the value size it represent.
    * Commonly used to compare the values of a variable at a given point in time.
    

```python
import matplotlib.pyplot as plt # import the scripting interface
import pandas as pd
import numpy as np
plt.style.use( {'dark_background','seaborn-v0_8-whitegrid'})

df_can = pd.read_csv("data/Canada.csv", index_col="Country")
years = list(map(str, range(1980,2014)))

df_can
```

```python
# Area Plots - Top 5 Countries with the highest Immigration to Canada

df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)
df_top5 = df_can.head(5)[years] # we only need the top 5 row data from each years per country
df_top5 = df_top5.transpose()
df_top5.rename(columns={"United Kingdom of Great Britain and Northern Ireland":"Great Britain"}, inplace=True)
print(df_top5.head())

df_top5.plot(kind="area", figsize=(12, 6))
plt.title("Immigration trend of top 5 countries")
plt.ylabel("Number of immigrants")
plt.xlabel("Years")


plt.show()
```

```python
# Histogram Plots - Number of Immigrants from 195 Countries to Canada in 2013

df_2013 = df_can.sort_index()["2013"]

df_2013.plot(kind="hist", figsize=(12, 6))
plt.title("Histogram of Immigration from 195 countries in 2013")
plt.ylabel("Number of Countries")
plt.xlabel("Number of Immigrants")

plt.show()
```

```python
# Note in the above figure x-axis, the markers are not align to the edge of the bars. 
# To correct the alignment, we use numpy to help us with the binning.

count, bin_edges = np.histogram(df_2013, bins=10)
df_2013.plot(kind="hist", xticks=bin_edges, figsize=(12, 6))

plt.title("Histogram of Immigration from 195 countries in 2013")
plt.ylabel("Number of Countries")
plt.xlabel("Number of Immigrants")

plt.show()
```

```python
# Generating a Bar Chart - Immigration from Iceland to Canada from 1980 to 2013

df_iceland = df_can.loc["Iceland", years]

df_iceland.plot(kind="bar", figsize=(12,6))

plt.title("Icelandic immigrants to Canada from 1980 to 2013")
plt.ylabel("Number of Immigrants")
plt.xlabel("Year")

plt.show()
```

```python
# Horizontal Bar Chart

df_iceland.plot(kind="barh", figsize=(12,6))

plt.title("Icelandic immigrants to Canada from 1980 to 2013")
plt.xlabel("Number of Immigrants")
plt.ylabel("Year")

plt.show()
```

```python
# Bar Chart with Highlighted Color

c = ['blue','red','blue','blue','blue','blue','blue','blue','blue','blue','red'] # bar color cycle
df_iceland.plot(kind="bar", figsize=(12,6), color = c, edgecolor='grey')

plt.title("Icelandic immigrants to Canada from 1980 to 2013")
plt.ylabel("Number of Immigrants")
plt.xlabel("Year")

plt.show()
```
