## Specialized Visualization Tools

### Pie Charts
* A pie chart is a Circular statistical graphic, divided into segments to illustrate numerical proportion.
* use the "explode" property to separate a specific segment from the pie chart.
* Bar chart communicates better than Pie chart.


```python
import matplotlib.pyplot as plt # import the scripting interface
import pandas as pd
import numpy as np
plt.style.use( {'seaborn-v0_8-dark-palette', 'seaborn-v0_8-darkgrid', 'seaborn-v0_8-paper'})

df_can = pd.read_csv("data/Canada.csv", index_col="Country")
years = list(map(str, range(1980,2014)))

df_can
```

Let's visualize the continent-wise breakdown of Immigration to Cananda from our dataset.
1. Group the data by Continent.
2. Plot the Pie Chart.


```python
df_continents = df_can.groupby('Continent').sum()
df_continents['Total'].plot(kind="pie", figsize=(10,6))

plt.title("Immigration to Canada by Continent [1980-2013]")
plt.show()
```

#### Box Plot
* A box plot is a way of statistically representing the distribution of data through five primary dimensions.
    * Minimum is usually the value obtained by subtracting 1.5 times the interquartile range, or IQR, from the first quartile.
    * First Quartile is the point 25% of the way through the sorted data.
    * Median is the median of the sorted data.
    * Third Quartile is the point 75% of the way through the sorted data.
    * Maximum is usually the value obtained by adding the third quartile to 1.5 times the IQR.
* Box plot also display outliers as individual dots outside the outer and lower extremes.


Create a Box Plot to visualize Immigration from Japan to Canada.

```python
df_japan = df_can.loc[['Japan'], years].transpose()
df_japan.plot(kind="box", figsize=(12,8))

plt.title("Box Plot of Japanese Immigrants from 1980-2013")
plt.ylabel("Number of Immigrants")
plt.show()
```

### Scatter Plot
* Display values pertaining to two variables.
* Determines the correlation between the two variables.
* Note with Scatter Plot we always plot between values of two columns from a DataFrame. Avoid using index.

```python
# Scatter Plot - Total Immigration to Canada from 1980 to 2013

df_total = pd.DataFrame(df_can[years].sum(), columns=["total"])
df_total.index.name = "year"
df_total = df_total.reset_index()

df_total.plot(kind="scatter", x='year', y='total', figsize=(12,8))
plt.title("Total Immigrant population to Canada from 1980-2013")
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")
plt.show()
```
