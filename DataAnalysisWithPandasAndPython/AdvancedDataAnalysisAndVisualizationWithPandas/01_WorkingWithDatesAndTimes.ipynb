## Working with Dates and Times

```python
import pandas as pd
import datetime as dt
```

### Review of Python's datetime Module
    * the datetime module is built into the core Python programming language.
    * the common alias for the datetime module is dt.
    * a module is a Python source file, think of it like an internal library that Python loads on demand.
    * the datetime module includes date and datetime classes for representing dates and datetimes.
    * the date constructor accepts arguments for year, month and day. Python defaults to 0 for any missing value.
    * the datetime constructor accepts arguments for year, month, day, hour, minute, and second.

```python
someday = dt.date(2025, 12, 15)
someday
```

```python
someday.year
```

```python
somedate = dt.datetime(2025,5,12, 16, 22, 34)
somedate
```

```python
somedate.month
```

```python
somedate.hour
```

### Create Range of Dates with pd.date_range Function
    * the date_range function generates and returns a DatetimeIndex holding a sequence of dates.
    * the function requires 2 of the 3 following parameters: start, end, and period.
    * with start and end, pandas will assume a daily period/interval.
    * every element within a DatetimeIndex is a Timestamp.

```python
pd.date_range(start="2026-01-01", end="2027-01-31") # default interval is 1 day
```

```python
pd.date_range(start="2026-01-01", end="2027-01-31", freq="2D") # interval of 2 day
```

```python
pd.date_range(start="2026-01-01", end="2027-01-31", freq="B") # interval of 1 Business Day
```

```python
pd.date_range(start="2026-01-01", end="2027-01-31", freq="W") # interval of Weekly
```

```python
pd.date_range(start="2026-01-01", end="2027-01-31", freq="W-FRI") # interval of Weekly Friday
```

```python
pd.date_range(start="2026-01-01", end="2027-01-31", freq="h") # Hourly interval
```

```python
pd.date_range(start="2026-01-01", end="2027-01-31", freq="6h") # 6 Hourly interval
```

```python
pd.date_range(start="2026-01-01", end="2027-01-31", freq="ME") # End of Monthly interval
```

```python
pd.date_range(start="2026-01-01", end="2027-01-31", freq="MS") # Start of Monthly interval
```

```python
pd.date_range(start="2026-01-01", end="2030-01-31", freq="YS") # Start of Annually interval
```

```python
pd.date_range(start="2026-01-01", end="2030-01-31", freq="YE") # End of Annually interval
```

```python
pd.date_range(start="2026-01-01", freq="MS", periods=24) # Start of Monthly interval for 24 periods.
```

```python
pd.date_range(end="2050-12-31", freq="ME", periods=24) # Month-Endly interval for 24 periods to end date.
```

### The dt Attribute
    * the dt attribute reveals a DatetimeProperties object with attributes/methods for working with datetimes. It is similar to the str attribute for string methods.
    * the DatetimeProperties object has attributes like day, month, and year to reveal information about each date in the Series.
    * the day_name method returns the written day of the week.
    * attributes like is_month_end and is_quarter_start return Boolean Series.

```python
ds = pd.Series(pd.date_range(start="2026-01-01", end="2030-01-31", freq="24D 3h")) # interval of 24 days 3 hours
ds
```

```python
ds.dt.day
```

```python
ds.dt.month
```

```python
ds.dt.year
```

```python
ds.dt.hour
```

```python
ds.dt.day_of_year
```

```python
ds.dt.day_name()
```

```python
ds.dt.day_of_week
```

```python
ds.dt.month_name()
```

```python
ds.dt.is_month_end
```

```python
ds[ds.dt.is_month_end]
```

### Selecting Rows from a DataFrame with a DateTimeIndex
    * the iloc accessor is available for index position-based extraction.
    * the loc accessor accepts strings or Timestamps to extract by index label/value. Note that Python's datetime objects will not work.
    * use list slicing to extract a sequence of dates. The truncate method is another alternative.

```python
ibm = pd.read_csv("data/ibm.csv", parse_dates=["Date"], index_col="Date").sort_index()
ibm
```

### The DateOffset Object
    * a DateOffset object adds time to a Timestamp to arrive at a new Timestamp.
    * the DateOffset constructor accepts days, weeks, months, years parameters, and more.
    * we can pass a DateOffset object to the freq parameter of the pd.date_range function.

```python
ibm
```

```python
ibm_idx = ibm.index
ibm_idx
```

```python
ibm_idx + pd.DateOffset(days=5)
```

```python
ibm_idx - pd.DateOffset(days=5)
```

```python
ibm_idx + pd.DateOffset(months=3)
```

```python
ibm_idx + pd.DateOffset(months=3, days=2)
```

```python
# find the ibm stock price on every one of my birthdays
birthdays = pd.date_range(start="1960-06-14", end="2023-06-14", freq=pd.DateOffset(years=1))
ibm[ibm.index.isin(birthdays)]
```

### Specialized Date Offsets
    * pandas nests more specialized date offsets in pd.tseries.offsets.
    * we can add a different amount of time to each date (e.g. month end, quarter end, year begin)

```python
ibm
```

```python
ibm.index
```

```python
# rounding the date to the end of month with specialized offsets object
ibm.index + pd.tseries.offsets.MonthEnd()
```

```python
# rounding the date to the end of previous month
ibm.index - pd.tseries.offsets.MonthEnd()
```

```python
# move date to next end quarter
ibm.index + pd.tseries.offsets.QuarterEnd()
```

```python
# move date to previous end quarter
ibm.index - pd.tseries.offsets.QuarterEnd()
```

```python
# move date to next beginning quarter, with starting month as Jan. Default is 3 or Mar
ibm.index + pd.tseries.offsets.QuarterBegin(startingMonth=1)
```

```python
# move date to next beginning quarter, with default starting month. Default is 3 or Mar
ibm.index + pd.tseries.offsets.QuarterBegin()
```

```python
# move date to previous beginning quarter, with starting month as Jan. Default is 3 or Mar
ibm.index - pd.tseries.offsets.QuarterBegin(startingMonth=1)
```

```python
# move date to next beginning of year
ibm.index + pd.tseries.offsets.YearBegin()
```

```python
# move date to previous beginning of year
ibm.index - pd.tseries.offsets.YearBegin()
```

### Timedeltas
    * a Timedelta is a pandas object that represents a duration (an amount of time).
    * subtracting two Timestamp objects will yield a Timedelta object (this applies to subtracting a Series from another Series).
    * the Timedelta constructor accepts parameters for time as well as string descriptions.

```python
pd.Timestamp("2023-03-31") - pd.Timestamp("2023-03-20")
```

```python
pd.Timestamp("2023-03-31 12:30:22") - pd.Timestamp("2023-03-20 19:24:55")
```

```python
pd.Timestamp("2023-03-20 19:24:55") - pd.Timestamp("2023-03-31 12:30:22")
```

```python
pd.Timedelta(days=3, hours=2, minutes=5)
```

```python
pd.Timedelta("5 minutes")
```

```python
pd.Timedelta("3 days 2 hours 5 minutes")
```
