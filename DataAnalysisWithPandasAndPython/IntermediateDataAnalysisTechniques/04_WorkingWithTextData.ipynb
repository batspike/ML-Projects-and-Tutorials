## Common String Methods
    * a Series has a special str attribute that expose an object with string methods.
    * Access the str attribute, then invoke the string method on the nested object.
    * Most method names will match their Python method equivalents (e.g. upper, lower, title, etc)

```python
chicago["Position Title"].str.lower()
# must use .str to apply string methods to a Series of strings
```

```python
chicago["Department"].str.replace("MGMNT", "MANAGEMENT").str.title()
# can chain string methods together to perform multiple operations at once
```

## Filtering with String Methods
    * the str.contains method checks whether a substring exists anywhere in the string
    * the str.startwith method checks whether a substring exists at the start of the string.
    * the str.endwith method checks whether a substring exists at the end of the string.

```python
water_workers = chicago["Position Title"].str.lower().str.contains("water")
chicago[water_workers].head()
```

```python
starts_with_civil = chicago["Position Title"].str.lower().str.startswith("civil")
chicago[starts_with_civil].head()
```

```python
ends_with_iv = chicago["Position Title"].str.lower().str.endswith("iv")
chicago[ends_with_iv].head()
```

## String Methods on Index and Columns
    * use the index and columns attributes to access the DataFrame index/column labels.
    * these objects support string methods via their own str attribute.

```python
chicago = pd.read_csv('data/chicago.csv', index_col="Name").dropna(how='all').sort_index()
chicago["Department"] = chicago["Department"].astype("category")
chicago
```

```python
chicago.index
```

```python
chicago.index = chicago.index.str.strip().str.title()
chicago.head()
```

```python
chicago.columns = chicago.columns.str.upper()
chicago
```

## The split Method
    * the str.split method splits a string by the occurrence of a delimiter. Pandas returns a Series of lists.
    * use the str.get method to access a nested list element by its index position.

```python
chicago = pd.read_csv('data/chicago.csv', index_col="Name").dropna(how='all').sort_index()
chicago["Department"] = chicago["Department"].astype("category")
chicago
```

```python
# The most common first word in "Position Title"

chicago["Position Title"].str.split(" ").str.get(0).value_counts()
# can split strings into lists and then use .str.get() to access specific elements of the list
```

```python
# Finding the most common first name among the employees in the dataset
chicago.index.str.title().str.split(", ").str.get(1).str.strip().str.split(" ").str.get(0).value_counts()
```

## The expand and n Parameters of the split Method
    * the expand parameter returns a DataFrame instead of a Series of lists.
    * the n parameter limits the number of splits.

```python
chicago.reset_index(drop=False, inplace=True)
chicago.head()
```

```python
# we specify a 2 columns list on the left hand side because the right hand side produces a DataFrame with 2 columns (the result of str.split with expand=True)
chicago[["Last Name", "First Name"]] = chicago["Name"].str.split(",", expand=True)
chicago
```

```python
# Splitting and expanding for n number of times
chicago[["Primary Title", "Secondary Title"]] = chicago["Position Title"].str.split(" ", expand=True, n=1)
chicago
```
