## Create a Multi Index
    * a multi index is an index with multiple levels or layers.
    * pass the set_index method a list of column names to create a multi-index DataFrame.
    * the order of the list's values will determine the order of the levels.
    * alternatively, we can pass the read_csv function's index_col parameter a list of columns.

## Extract Index Level Values
    * the get_level_values method extracts an index with the values from one level in the MultiIndex.
    * invoke the get_level_values on the MultiIndex, not the DataFrame itself.
    * the method expects either the level's index position or its name.

```python
bigmac.index.get_level_values("Date")
# returns the values of the Date level of the index
```

```python
bigmac.index.get_level_values(0)
# returns the values of the first level of the index, which is Date in this case
```

```python
bigmac.index.get_level_values("Country")
```

```python
bigmac.index.get_level_values(1)
```

## Rename Index Levels
    * invoke the set_names method on the MultiIndex to change one or more level names.
    * use the names and level parameter to target a nested index at a given level.
    * alternatively, pass names a list of strings to overwrite all level names.
    * the set_names method returns a copy, so replace the original index to alter the DataFrame.

```python
bigmac.index.set_names(names="Time", level=0, inplace=True)
bigmac.head()
```

```python
bigmac.index.set_names(names="Location", level=1, inplace=True)
bigmac.head()
```

## The sort_index Method on a Multi Index DataFrame
    * using the sort_index method, we can target all levels or specific levels of the MultiIndex.
    * to apply a different sort order to different levels, pass a list of Booleans.

```python
bigmac = pd.read_csv('data/bigmac.csv', parse_dates=['Date'], date_format='%Y-%m-%d', index_col=["Date","Country"])
bigmac
```

```python
bigmac.sort_index(inplace=False)
# sorts the index from the least significant level to the most significant level, which is the opposite of how the data is sorted in the file. 
# This is because the index is sorted in ascending order by default, and the most significant level 0 is sorted first.   
```

```python
bigmac.sort_index(ascending=False,inplace=False)
# sorts the index in descending order, which is the same as how the data is sorted in the file.
```

```python
bigmac.sort_index(ascending=[False, True],inplace=False)
# sorts the index in descending order by the first level and in ascending order by the second level.
```

## Extract Rows from a Multiindex DataFrame
    * a tuple is an immutable list. It cannot be modified after creation.
    * create tuple with a comma between elements. The community convention is to wrap the elements in parenthesis.
    * the iloc and loc accessors are available to extract rows by index position or label.
    * for the loc accessor, pass a tuple to hold the labels from the index levels.

```python
bigmac = pd.read_csv('data/bigmac.csv', parse_dates=['Date'], date_format='%Y-%m-%d', index_col=["Date","Country"]).sort_index()
bigmac
```

```python
# iloc ignores the levels and labels in Multi-index. iloc access by the literal row of the dataframe.
bigmac.iloc[0] # access the first row, 
# note the output shows the Series name as a tuple of the multi index.
```

```python
bigmac.iloc[1] # access the second row
```

```python
bigmac.iloc[1:4] # access a range from 2nd to 3rd row (slicing)
```

```python
# accessing a specific row and column, df.iloc[row_index, column_index]
bigmac.iloc[1,0] # we only have one column in our dataframe
```

```python
bigmac.iloc[1:3, 0] # row 2 to 3, column 1
```

```python
bigmac.iloc[[1,4,5]]  # accessing specific rows in a list
```

```python
# for loc[ (index label as a tuple) ] , we provide a tuple to the index
bigmac.loc[("2000-04-01", "Canada")] # access with key level 1 = "2000-04-01" and level 2 = "Canada"
```

```python
bigmac.loc[("2000-04-01")] # every rows where level 1 = "2000-04-01"
```

```python
# the second value in the tuple, can also be a column name !! not recommended as it is confusing.
bigmac.loc[ ("2000-04-01", "Price in US Dollars") ]
```

```python
# do this instead
bigmac.loc[ ("2000-04-01", ), "Price in US Dollars" ]
```

```python
# extract by range of multi index.
bigmac.loc[("2000-04-01","Hungary") : ("2012-01-01","UAE")]
```

```python
bigmac.loc[("2000-04-01","Hungary") : ]
```

```python
bigmac.loc[ : ("2012-01-01","UAE")]
```

```python
# all rows for "Canada" , ignoring the Date

idx = pd.IndexSlice
bigmac.loc[ idx[:,"Canada"], : ]
```

```python
# alternatively, - not the preferred method
bigmac.loc[ (slice(None), "Canada"), :]
```

```python

# using IndexSlice is more flexible, e.g.

idx = pd.IndexSlice
bigmac.loc[ idx[ "2012-01-01":"2018-01-01", ["Canada","UAE"]], : ]
```

## The stack Method
    * the stack method moves the column index to the row index.
    * pandas will return a MultiIndex Series.
    * think of it like "stacking" index levels for a MultiIndex.

```python
# we chose year as the outer index since it has the least number of unique values.
world = pd.read_csv('data/worldstats.csv', index_col=["year","country"]).sort_index()
world
```

```python
# stack() rotates the column labels into the row index.
# Before: we have a DataFrame with a 2-level index (year,country) and 2 columns (Population and GDP).
# After : we get a Series with a 3-level index (year,country,Population and GDP).
# We use stack for:
#   1. Reshaping for Plots: Plotting library prefer Long-Form data where every row is a single observation.
#   2. Uniform Analysis: to perform calculation that applies to both Population and GDP, it is easier when they are in the same column.
# You can undo it using unstack() method.
world.stack()
```

## The unstack Method
    * the unstack method moves a row index to the column index.
    * by default, the unstack method will move the innermost index.
    * we can customize the moved index with the level parameter.
    * the level parameter accepts the level's index position or its name. It can also accept a list of positions/names.

```python
world
```

```python
world_stack = world.stack()
world_stack
```

```python
world_stack.columns
```

```python
world_unstack = world_stack.unstack()
world_unstack
```

```python
world_unstack.columns
```

```python
world_unstack.unstack()
```

```python
world_unstack.unstack().columns
```

```python
world_stack.unstack(level=0)
```

```python
world_stack.unstack(level="year") # same as above
```

```python
world_stack.unstack(level=1)
```

```python
world_stack.unstack(level="country") # same as above
```

```python
world_stack.unstack(level=2)
```

```python
world_stack
```

```python
# unstacking with a list of column names
world_stack.unstack(["country","year"])
```

```python
world_stack.unstack([1,0]) # same as above
```

## The pivot Method
    * the pivot method reshapes data from a tall format to a wide format.
    * ask yourself which direction the data will expand in if you add more entries.
    * a tall/long foramt expands down. A wide format expands out.
    * the index parameter sets the horizontal index of the pivoted DataFrame.
    * the columns parameter sets the column whose values will be the columns in the pivoted DataFrame.
    * the values parameter set the values of the pivoted DataFrame. Pandas will populate the correct values based on the index and column intersections.

```python
sales = pd.read_csv("data/salesmen.csv")
sales
```

```python
sales_pivot = sales.pivot(index="Date", columns="Salesman", values="Revenue")
sales_pivot
```

```python
type(sales_pivot)
```

## The melt Method
    * the melt method is the inverse of the pivot method.
    * it takes a wide dataset and converts it to a tall dataset.
    * the melt method is ideal when you have multiple columns storing the same data point.
    * ask yourself whether the column's values are a type of the column header. If they are not, the data is likely stored in a wide format.
    * the id_vars parameter accepts the column whose values will be repeated for every column.
    * the var_name parameter sets the name of the new column for the varying values (the former column names).
    * the value_name parameter set the new name of the values column (holding the values from the original DataFrame).

```python
quarters = pd.read_csv("data/quarters.csv")
quarters
```

```python
quarters.melt(id_vars="Salesman", var_name="Quarter", value_name="Revenue")
```

## The pivot_table Method
    * the pivot_table method operates similarly to the Pivot Table feature in Excel.
    * a pivot table is a table whose values are aggregations of groups of values from another table.
    * the values parameter accepts the numeric column whose values will be aggregated.
    * the aggfunc parameter declares the aggregation function (the default is mean/average).
    * the index parameter sets the index labels of the pivot table. MultiIndexes are permitted.
    * the columns parameter sets the column labels of the pivot table. MultiIndexes are permitted.

```python
foods = pd.read_csv("data/foods.csv")
foods
```

```python
foods.pivot_table(values="Spend", index="Gender", aggfunc="mean")
```

```python
foods.pivot_table(values="Spend", index="Gender", aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index="Item", aggfunc="mean")
```

```python
foods.pivot_table(values="Spend", index=["Gender","Item"], aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index=["Gender","Item"], columns="City", aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index="Item", columns=["Gender","City"], aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index="Item", columns=["Gender","City"], aggfunc="count")
```

```python
foods.pivot_table(values="Spend", index="Item", columns=["Gender","City"], aggfunc="max")
```
