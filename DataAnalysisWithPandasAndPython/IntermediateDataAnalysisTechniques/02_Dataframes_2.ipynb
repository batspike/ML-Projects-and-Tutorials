## Data Types Conversion for Memory Optimization
    * the pd.to_datetime method converts a Series to hold datetime values.
    * the format parameter informs pandas of the format that the times are stored in.
    * we pass symbols designating the segments of the string. e.g. %m means 'month' and %d means 'day'.
    * the dt attribute reveals an object with many datetime related attributes and methods.
    * the dt.time attribute extracts only the time from each value in a datetime Series.
    * use the astype method to convert the values in a Series to another type.
    * the parse_dates parameter of read_csv is an alternate way to parse strings as datetimes.

```python
import pandas as pd

employees = pd.read_csv('data/employees.csv')
employees.info()
```

```python
employees
```

```python
# Convert the "Start Date" column to datetime format
employees["Start Date"] = pd.to_datetime(employees['Start Date'], format='%m/%d/%Y') 
employees
```

```python
# Convert the "Last Login Time" column to datetime format (time only)
employees["Last Login Time"] = pd.to_datetime(employees['Last Login Time'], format='%I:%M %p').dt.time
employees
```

```python
# Convert the "Senior Management" column to boolean type
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees
```

```python
# Convert the Gender column to category type
employees["Gender"] = employees["Gender"].astype('category')
employees.info()
```

```python
# alternatively, we can specify the data types when reading the CSV file
employees = pd.read_csv('data/employees.csv', parse_dates=['Start Date'], dtype={'Gender': 'category'})
employees["Last Login Time"] = pd.to_datetime(employees['Last Login Time'], format='%I:%M %p').dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees.info()
```

```python
employees
```

## Filter a DataFrame Based on Condition
    * Pandas needs a Series of Booleans to perform filtering.
    * Pass the Boolean Series inside square brackets after the DataFrame.
    * we can generate a Boolean Series using a wide variety of operation (equality, inequality, less than, greater than, etc)

```python
employees
```

```python
# Filtering for Male employees
male_employees = employees[employees["Gender"] == 'Male'] 
male_employees
```

```python
# Filtering for Finance department employees
finance_employees = employees[employees["Team"] == 'Finance']
finance_employees
```

```python
# Filtering for Senior Management employees
senior_management_employees = employees[employees["Senior Management"]]
senior_management_employees
```

```python
# Filtering for employees with Salary greater than $110,000
high_salary_employees = employees[employees["Salary"] > 110000]
high_salary_employees
```

```python
# Filtering for employees with less than 1.5% Bonus
low_bonus_employees = employees[employees["Bonus %"] < 1.5]
low_bonus_employees
```

```python
# Employees that started before 1985
early_start_employees = employees[employees["Start Date"] < "1985-01-01"]
early_start_employees
```

```python
import datetime as dt

# Employees that logged in before noon
morning_login_employees = employees[employees["Last Login Time"] < dt.time(12, 0)]
morning_login_employees
```

## Filter with More than One Condition (AND - &)
    * add the & operator in between two boolean Series to filter by multiple conditions.
    * we can assign the Series to variables to make the syntax more readable.

```python
# female employees who work in Marketing
is_female = employees["Gender"] == 'Female'
is_marketing = employees["Team"] == 'Marketing'
female_marketing_employees = employees[ is_female & is_marketing]
female_marketing_employees
```

```python
# female employees who work in Marketing who earn more than $100,000
is_high_earning = employees["Salary"] > 100000
female_high_earning_marketing_employees = employees[is_female & is_marketing & is_high_earning]
female_high_earning_marketing_employees
```

## Filter with More than One Condition ( OR  - | )
    * use the | operator in between two boolean Series to filter by either condition

```python
# Employees who are either Senior Management or started before 1990
is_senior_management = employees["Senior Management"]
started_before_1990 = employees["Start Date"] < "1990-01-01"
senior_or_early_employees = employees[is_senior_management | started_before_1990]
senior_or_early_employees
```

```python
# employees whose First Name is 'Robert' who work in Client Services or Started after 2016-06-01
is_named_robert = employees["First Name"] == 'Robert'
is_client_services = employees["Team"] == 'Client Services'
started_after_june_2016 = employees["Start Date"] > "2016-06-01"
robert_client_services_or_recent_employees = employees[(is_named_robert & is_client_services) | started_after_june_2016]
robert_client_services_or_recent_employees
```

## The isin() Method
    * the isin() Series method accepts a collection object like a list, tuple, or Series.
    * the method returns True for a row if its value is found in the collection.

```python
# employees who are on Legal team or in Sales team or in Product team
is_legal = employees["Team"] == 'Legal'
is_sales = employees["Team"] == 'Sales'
is_product = employees["Team"] == 'Product'
legal_sales_or_product_employees = employees[is_legal | is_sales | is_product]
legal_sales_or_product_employees
```

```python
is_legal_sales_or_product_employees = employees["Team"].isin(['Legal', 'Sales', 'Product'])
legal_sales_or_product_employees = employees[is_legal_sales_or_product_employees]
legal_sales_or_product_employees
```

## The isnull and notnull Methods
    * the isnull method returns True for NaN values in a Series
    * the notnull method returns True for present values in a Series

```python
# employees with missing or null Team information
missing_team_employees = employees[employees["Team"].isnull()]
missing_team_employees
```

```python
# employees who are not null in the Team column
non_null_team_employees = employees[employees["Team"].notnull()]
non_null_team_employees
```

## The between Method
    * the between method returns True if a Series value is found within its range

```python
# employees whose salary is between $90,000 and $120,000
is_salary_between_90k_and_120k = employees["Salary"].between(90000, 120000)
salary_between_90k_and_120k_employees = employees[is_salary_between_90k_and_120k]
salary_between_90k_and_120k_employees
```

```python
# employees started between 1980 and 1990
is_started_between_1980_and_1990 = employees["Start Date"].between("1980-01-01", "1990-12-31")
started_between_1980_and_1990_employees = employees[is_started_between_1980_and_1990]
started_between_1980_and_1990_employees
```

```python
# employees who login in between 9:00 AM and 5:00 PM
is_login_between_9_and_5 = employees["Last Login Time"].between(dt.time(9, 0), dt.time(17, 0))
login_between_9_and_5_employees = employees[is_login_between_9_and_5]
login_between_9_and_5_employees
```

## The duplicated Method
    * the duplicated method returns True if a Series value is a duplicate.
    * Pandas will mark one occurance of a repeated value as a non-duplicate.
    * use the keep parameter to designate whether the first or last occurance of a repeated value should be considered the 'non-duplicate'.
    * pass False to the keep parameter to mark all occurances of repeated values as duplicates.
    * use the tilde symbol ~ to invert a Series's values. Trues will become Falses, and vice versa.

```python
# employees with duplicated First Name.
duplicated_first_names = employees[employees.duplicated(subset=["First Name"], keep=False)]
duplicated_first_names
```

```python
duplicated_first_names = employees[employees.duplicated(subset=["First Name"], keep='first')]
duplicated_first_names
```

```python
duplicated_first_names = employees[employees.duplicated(subset=["First Name"], keep='last')]
duplicated_first_names
```

```python
# inverse of duplicated: employees with unique First Name
duplicated_first_names = employees[~employees.duplicated(subset=["First Name"], keep=False)]
duplicated_first_names
```

## The drop_duplicates Method
    * the drop_duplicates method deletes rows with duplicate values.
    * by default, it will remove a row if all of its values are shared with another row.
    * the subset parameter configures the columns to look for duplicate values within.
    * pass a list to subset parameter to look for duplicates across multiple columns.

```python
employees
```

```python
employees.drop_duplicates()
```

```python
employees.drop_duplicates(subset=["Team"])
```

```python
employees.drop_duplicates(subset=["Team"], keep='first')
```

```python
employees.drop_duplicates(subset=["Team"], keep='last')
```

```python
employees.drop_duplicates(subset=["Team"], keep=False)
```

```python
employees.drop_duplicates(subset=["First Name"], keep=False)
```

```python
employees.drop_duplicates(subset=["Senior Management","Team"])
```
