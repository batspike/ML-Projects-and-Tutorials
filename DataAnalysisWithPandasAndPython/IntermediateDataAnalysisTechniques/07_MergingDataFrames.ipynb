## The pd.concat Function
    * the concat function concatenates one DataFrame to the end of another.
    * the original index labels will be kept by default. Set ignore_index to True to generate a new index.
    * the keys parameter create a MultiIndex using the specified keys/labels.
    * pandas will concatenate the DataFrames along the row/index axis.
    * pandas will include all columns that exist in either DataFrame. If there are no matching values, pandas will use NaN values.
    * we can pass the axis parameter an argument of "columns" to concatenate on the column axis.

```python
pd.concat([week1, week2])
```

```python
pd.concat([week1, week2], ignore_index=True)
```

```python
pd.concat([week1, week2], keys=["Week 1","Week 2"]) # returns a multi index dataframe
```

```python
df1 = pd.DataFrame([1,2,3], columns=["A"])
df2 = pd.DataFrame([1,2,3], columns=["B"])

pd.concat([df1,df2])
# since the column labels in the two df are different, pandas will simply append the df2 to df1 with two different columns.
# this default behaviour is merging along the index axis.
```

```python
pd.concat([df1,df2],axis="index") # same as the above, default behaviour is merge along the index axis, i.e. vertically
```

```python
pd.concat([df1,df2], axis="columns") # merge along the column axis, i.e. horizontally
```

```python
df3 = df2.rename({"B":"A"}, axis=1)
df3
```

```python
pd.concat([df1,df3]) # default merging behaviour where the two dataframes have the same column labels.
```

```python
pd.concat([df1,df3], axis="index") # same as above
```

```python
pd.concat([df1,df3], axis="columns") # merge horizontally
```

```python
pd.concat([df1,df3], ignore_index=True) # merge vertically and reindex
```

## Left Joins
    * the merge method joins two DataFrames together based on shared values in a column or an index.
    * a left join merges one DataFrame into another based on values in the first one.
    * the left DataFrame is the one we invoke the merge method on.
    * if the left DataFrame's value is not found in the right DataFrame, the row will hold NaN values.

```python
week1
```

```python
foods
```

```python
# for each customer transaction in week1, which food item did the customer ordered
week1.merge(right=foods, how="left", on="Food ID")
```

```python
# for each food item, which customer order it in week1
foods.merge(right=week1, how="left", on="Food ID")
```

```python
# Food Item order count in week1
foods.merge(right=week1, how="left", on="Food ID").value_counts("Food Item")
```

## The left_on and right_on Parameters
    * the left_on and right_on parameters designate the column names from each DataFrame to use in the merge.
    * this means we need not rename the join column names.

```python
week2.head()
```

```python
customers.head()
```

```python
week2.merge(customers, how="left", left_on="Customer ID", right_on="ID")
# note the right join column ("ID") is not shown in the result as it is the index of the right table.
```

## Inner Joins
    * inner joins merge two tables based on shared/common values in columns.
    * if only one DataFrame has a value, pandas will exclude it from the final results set.
    * if the same ID occurs multiple times, pandas will store each possible combination of the values.
    * the design of the join ensures that the results will be the same no matter wheat DataFrame the merge method is invoked upon.
    * we can pass multiple arguments to the on parameter of the merge method. Pandas will require matches in both columns across the DataFrames.

```python
week1.head()
```

```python
week2.head()
```

```python
# the customers that came in in both weeks
week1.merge(week2, how="inner", on="Customer ID")
```

```python
t = week1.merge(week2, how="inner", on="Customer ID", suffixes=[" - Week 1", " - Week 2"])
t
```

```python
# multiple rows with same Customer ID in week 1
week1[week1["Customer ID"] == 77]
```

```python
week2[week2["Customer ID"] == 77]
```

```python
# all possible join result for the same Customer ID

t[t["Customer ID"] == 77]
```

```python
# the "on" parameter

week1
```

```python
week2
```

```python
# customers that came in on both weeks and order the same food
week1.merge(week2, how="inner", on=["Customer ID", "Food ID"], suffixes=[" - Week 1", " - Week 2"]) 
# suffixes is not needed since both table has same column names.
```

```python
week1[(week1["Customer ID"] == 21) & (week1["Food ID"] == 4)]
```

```python
week2[(week2["Customer ID"] == 21) & (week2["Food ID"] == 4)]
```

## Full-Outer Joins
    * a full/outer joins values that are found in either DataFrame or both DataFrames.
    * pandas does not mind if a value exists in one DataFrame but not the other.
    * if a value does not exist in one DataFrame, it will have a NaN.

```python
week1
```

```python
week2
```

```python
week1.merge(week2, how="outer", on="Customer ID", suffixes=[" - Week 1", " - Week 2"])
```

```python
# same as above with indicator flag
week1.merge(week2, how="outer", on="Customer ID", suffixes=[" - Week 1", " - Week 2"], indicator=True)
```

```python
# same as above with custom label
t = week1.merge(week2, how="outer", on="Customer ID", suffixes=[" - Week 1", " - Week 2"], indicator="Merge Indicator")
t.head(20)
```

```python
t["Merge Indicator"].value_counts()
```

```python
t[t["Merge Indicator"].isin(["left_only","right_only"])].head(20)
```

## Merging by Indexes with the left_index and right_index Parameters
    * use the on parameter if the column(s) to be matched on have the same names in both DataFrames.
    * use the left_on and right_on parameters if the column(s) to be matched on have different names in the two DataFrames.
    * use the left_index or right_index parameters (set to True) if the values to be matched on are found in the index of a DataFrame.

```python
week1
```

```python
customers
```

```python
week1.merge(customers, how="left", left_on="Customer ID", right_index=True).merge(foods, how="left", left_on="Food ID", right_index=True)
```
