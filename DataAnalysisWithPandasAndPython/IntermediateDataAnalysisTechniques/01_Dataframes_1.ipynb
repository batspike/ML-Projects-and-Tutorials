## Methods and Attributes between Series and DataFrames
#### Series and DataFrames have many methods and attributes in common, but they also have some that are unique to each type. Here are some of the key differences:
#### 1. Methods:
* Series:
    * `Series.mean()`: Calculates the mean of the values in the Series.
    * `Series.median()`: Calculates the median of the values in the Series.
    * `Series.mode()`: Calculates the mode of the values in the Series.
    * `Series.value_counts()`: Counts the unique values in the Series and returns a Series with the counts.
    * `Series.unique()`: Returns the unique values in the Series.
* DataFrame:
    * `DataFrame.mean()`: Calculates the mean of the values in each column of the DataFrame.
    * `DataFrame.median()`: Calculates the median of the values in each column of the DataFrame.
    * `DataFrame.mode()`: Calculates the mode of the values in each column of the DataFrame.
    * `DataFrame.value_counts()`: Counts the unique values in each column of the DataFrame and returns a DataFrame with the counts.
    * `DataFrame.unique()`: Returns the unique values in each column of the DataFrame.
#### 2. Attributes:
* Series:
    * `Series.name`: The name of the Series.
    * `Series.index`: The index of the Series.
    * `Series.dtype`: The data type of the values in the Series.
* DataFrame:
    * `DataFrame.columns`: The column names of the DataFrame.
    * `DataFrame.index`: The index of the DataFrame.
    * `DataFrame.dtypes`: The data types of the columns in the DataFrame.
    * `DataFrame.shape`: The shape of the DataFrame (number of rows and columns).
    * `DataFrame.info()`: Provides a summary of the DataFrame, including the number of non-null values, data types, and memory usage.
 
 These are just a few examples of the methods and attributes that are unique to Series and DataFrames. 
 Both Series and DataFrames have many other methods and attributes that can be used for data manipulation and analysis.

## Differences between Shared Methods
    * The sum method adds a Series values
    * On a Dataframe, the sum method defaults to adding the values by traversing the index (row values)
    * The axis parameter customizes the direction that we add across. Pass "columns" or 1 to add "across" the columns

```python
revenue = pd.read_csv("data/revenue.csv", index_col="Date") # Reads a CSV file and sets the "Date" column as the index of the resulting DataFrame.
revenue
```

```python
s = pd.Series([1, 2, 3])
s.sum()
```

```python
# A Series is 1D and a sum() method automatically sums up the values in one direction (rows), while a DataFrame is 2D and a sum() method 
# sums up the values in both directions (columns and rows) by default. Both Series and DataFrame will traverse the values on its index and 
# sum up the values in the column. Since a Series has only one column, it will just sum up that column. On the other hand, a DataFrame has 
# multiple columns, so it will sum up each column separately and return a Series with the sums of each column.
revenue.sum() # Sums up the values in each column of the DataFrame.
```

```python
revenue.sum().sum() # sums up everything in the DataFrame, first summing each column and then summing those sums together.
```

```python
# Since DataFrame has multiple columns, the sum() method also provides an option to specify the axis along which to perform the summation. 
# By default, it sums along the columns (axis=0), but you can also sum along the rows (axis=1) by passing the appropriate argument to the sum() method.
revenue.sum(axis=1) # Sums up the values in each row of the DataFrame, i.e., it sums across the columns of New York, San Francisco, and Los Angeles for each row.
```

```python
revenue.sum(axis="columns").sum() # Sums up the values in each row of the DataFrame, using the string "columns" as an alternative to axis=1.
```

## Select One Column from a Dataframe

    * use square bracket syntax (df["column_name"]) which works with spaces.
    * Pandas extracts a column from a Dataframe as a Series.
    * The Series is a view, however as of Pandas 3.0, modifying the Series will result in a copy and the changes will NOT affects the original Dataframe.


```python
nba
```

```python
nba["Team"] # Another way to access the "Team" column of the DataFrame, which also returns a Series containing the team names.
```

## Drop DataFrame Rows with Missing Values
    * Pandas uses a NaN designation for cells that have a missing value.
    * The dropna method deletes rows with missing values. Its default behaviour is to remove a row if it has any missing values.
    * Pass the how parameter an argument of "all" to delete rows where all the values are NaN.
    * The subset parameters customizes/limits the columns that pandas will use to drop rows with missing values.

```python
nba
```

```python
nba.dropna() # by default, it will drop rows where any column has a missing value (NaN). If you want to drop columns instead, you can use the axis parameter and set it to 1 (axis=1) or 'columns' (axis='columns').  
```

```python
nba.dropna(how="any") # Drops rows where any column has a missing value (NaN). This is the default behavior of dropna().
```

```python
nba.dropna(how="all") # Drops rows where all columns have missing values (NaN). The how parameter with the value "all" specifies that only rows where all values are NaN should be dropped.
```

```python
nba.dropna(subset=["College"]) # Drops rows where the "College" column has missing values (NaN). The subset parameter specifies that only the "College" column should be considered when determining which rows to drop based on missing values.   
 
```

```python
nba.dropna(subset=["College", "Salary"]) # Drops rows where the "College" or "Salary" column has a missing value (NaN). The subset parameter specifies that only the "College" and "Salary" columns should be considered when determining which rows to drop based on missing values.  
```

```python
nba.dropna(subset=["College", "Salary"], how="all") # drop the row only when both "College" and "Salary" columns have missing values (NaN).
# Drops rows where the "College" or "Salary" column has a missing value (NaN). 
# The subset parameter specifies that only the "College" and "Salary" columns should be considered when determining which rows to drop based on missing values. 
# The how parameter with the value "all" specifies that only rows where all values in the specified subset are NaN should be dropped.
```

## Fill in Missing Values with the fillna Method
    * the fillna method replaces missing NaN values with its argument.
    * the fillna method is available on both DataFrame and Series.
    * an extracted Series is a view of the original DataFrame, but the fillna method returns a copy.

```python
nba.dropna(how="all", inplace=True)
nba
```

```python
nba.fillna({"Salary": 0}) # Fills missing values (NaN) in the "Salary" column with the value 0. 
# The fillna() method is used to replace missing values in a DataFrame, and by specifying the column name and the value to fill, it will only fill the missing values in that specific column.
```

## The astype() Method
    * the astype() converts a Series's values to a specified type.
    * pass in the specified type as either a string or the core Python data type.
    * Pandas cannot convert NaN values to numeric types, so we need to eliminate/replace them before we perform the conversion.
    * the dtypes attribute returns a Series with the DataFrame's columns and their types.

    * the category type is ideal for columns with a limited number of unique values.
    * the nunique method will return a Series with the number of unique values in each column.
    * with categories, pandas does not create a separate value in memory for each 'cell'. Rather the cell's point to a single copy for each unique value.

```python
nba
# whenever pandas import a CSV file, it will automatically convert empty cells to NaN (Not a Number) to represent missing values. 
# Also, as long as there is a NaN value in a column, the column will be converted to a float data type, even if the other values in the column are integers. 
# This is because NaN is a special floating-point value, and pandas promotes the entire column to a float data type to accommodate the presence of NaN values.
# This allows for easier handling of missing data in the DataFrame, as you can use methods like dropna() and fillna() to manage those missing values effectively.
```

```python
nba.dtypes
```

```python
nba['Salary'] = nba['Salary'].astype("Int64") 
# Converts the "Salary" column to the nullable integer data type "Int64". 
# This allows for the presence of missing values (NaN) while still treating the column as integers. The "Int64" data type is a special pandas extension type 
# that can handle missing values without converting the entire column to a float data type.

nba
```

```python
nba.dtypes
```

```python
nba['Weight'] = nba['Weight'].fillna(0).astype(int) 
# To convert a type, none of the column values can be NaN.
# Fills missing values (NaN) in the "Weight" column with 0 and then converts the column to the integer data type. 
# This allows for the presence of missing values (NaN) while still treating the column as integers. The fillna(0) method replaces NaN values with 0, 
# and then astype(int) converts the entire column to integers.
```

```python
nba.dtypes
```

```python
nba['Team'].value_counts()
```

```python
nba['Team'].nunique() # Returns the number of unique values in the "Team" column. The nunique() method counts the number of distinct values in a Series or DataFrame column.
```

```python
nba.nunique() # Returns the number of unique values in each column of the DataFrame. The nunique() method counts the number of distinct values in each column and returns a Series with the counts for each column.
```

```python
nba.info()
```

```python
nba['Position'] = nba['Position'].astype("category") # Converts the "Position" column to the categorical data type. This can be useful for columns that have a limited number of unique values, as it can save memory and improve performance when performing operations on that column.
nba.dtypes
```

## Sort a DataFrame with the sort_values() Method
    * the sort_values method sorts a DataFrame by the values in one or more columns. The default is in ascending order.
    * the first parameter 'by' expects the column(s) to sort by.
    * if sorting by a single column, pass a string with the column name.
    * the ascending parameter specify the sort order.
    * the na_position parameter customizes where pandas places NaN value.

    * to sort multiple columns, pass the by parameter a list of column names. Pandas will sort in the specified column order (first to last).
    * pass the ascending parameter a boolean to sort all columns in a consistent order (all ascending or all descending).
    * pass ascending a list of customize sort order per column. the ascending list length must match the by list.

```python
nba = pd.read_csv("data/nba.csv")
nba
```

```python
nba['Name'].sort_values() # Sorts the values in the "Name" column in ascending order. The sort_values() method sorts the values in a Series or DataFrame column and returns a new Series with the sorted values.
```

```python
nba['Name'].sort_values(ascending=False) # Sorts the values in the "Name" column in descending order. The sort_values() method with the ascending=False argument sorts the values in descending order and returns a new Series with the sorted values.
```

```python
nba['Salary'].sort_values() # Sorts the values in the "Salary" column in ascending order. The sort_values() method sorts the values in a Series or DataFrame column and returns a new Series with the sorted values.
```

```python
nba['Salary'].sort_values(na_position="first") # Sorts the values in the "Salary" column in ascending order, placing any missing values (NaN) at the beginning of the sorted output. The na_position parameter with the value "first" specifies that NaN values should be placed at the beginning of the sorted output.
```

```python
nba[["Name", "Salary"]].sort_values(by="Salary") # Sorts the DataFrame by the "Salary" column in ascending order, returning a new DataFrame with the sorted values. The sort_values() method with the by argument specifies which column to sort by.
```

```python
nba[["Name", "Salary"]].sort_values(by="Name", ascending=False) # Sorts the DataFrame by the "Name" column in descending order, returning a new DataFrame with the sorted values. The sort_values() method with the by argument specifies which column to sort by, and the ascending=False argument sorts the values in descending order.
```

```python
nba[["Name", "Salary"]].sort_values(by="Salary", ascending=False) # Sorts the DataFrame by the "Salary" column in descending order, returning a new DataFrame with the sorted values. The sort_values() method with the by argument specifies which column to sort by, and the ascending=False argument sorts the values in descending order.
```

```python
nba[["Position", "Salary"]].sort_values(by=["Position","Salary"])
```

```python
nba[["Position", "Salary"]].sort_values(by=["Position","Salary"], ascending=[True, False]) # Sorts the DataFrame first by the "Position" column in ascending order and then by the "Salary" column in descending order. The sort_values() method with the by argument specifies which columns to sort by, and the ascending argument with a list of boolean values specifies the sorting order for each column.
```

## Sort DataFrame with the rank() Method
    * the rank() method assigns a numeric ranking to each Series value.
    * Pandas will assign the same rank to equal values and create a 'gap' in the dataset for the ranks.

```python
nba = pd.read_csv("data/nba.csv").dropna(how="all") # Reads the CSV file and drops rows where all columns have missing values (NaN). The dropna() method with the how="all" argument specifies that only rows where all values are NaN should be dropped.
nba = nba.fillna(0)
nba
```

```python
nba["Salary"].rank() # Returns the rank of each value in the DataFrame, with the default method of ranking being "average". The rank() method assigns ranks to values in a Series or DataFrame column, with the option to specify different ranking methods.
```

```python
nba["Salary"].rank(ascending=False) .astype(int) # Returns the rank of each value in the "Salary" column in descending order, and then converts the ranks to integers. The rank() method with the ascending=False argument assigns ranks in descending order, and the astype(int) method converts the resulting ranks to integers.

```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype(int) # Adds a new column "Salary Rank" to the DataFrame, which contains the rank of each value in the "Salary" column in descending order, with the ranks converted to integers. The rank() method with the ascending=False argument assigns ranks in descending order, and the astype(int) method converts the resulting ranks to integers.
nba
```

```python
nba.sort_values(by="Salary", ascending=False) # Sorts the DataFrame by the "Salary" column in descending order, returning a new DataFrame with the sorted values. The sort_values() method with the by argument specifies which column to sort by, and the ascending=False argument sorts the values in descending order.
```
