## Module's Dataset
    * this modul's dataset is a collection of all the James Bond movies.

```python
import pandas as pd

bond = pd.read_csv('data/jamesbond.csv')
bond
```

## The set_index and reset_index Methods
    * the index serves as the collection of primary identifiers/labels/entrypoints for the rows.
    * the fastest way to extract a row is from a sorted index by position/label.
    * Pandas uses index labels/values when merging different objects together.
    * the set_index method sets an existing column as the index of the DataFrame.
    * the reset_index method sets the standard ascending numeric index as the index of the DataFrame.

```python
bond.set_index('Film', inplace=True)
bond
```

```python
bond.reset_index(inplace=True) # resets the index to the default integer index and moves the 'Film' column back to a regular column
```

## Retrieve Rows by Index Position with iloc Acessor
    * the iloc accessor retrieves one or more rows by index position.
    * provide a pair of square brackets after the acessor.
    * iloc accepts single values, lists, and slices.

```python
bond.head()
```

```python
bond.iloc[5] # returns the 6th row of the dataframe, which is the film "Goldfinger"
```

```python
bond.iloc[3:15] # returns the 4th to 15th rows of the dataframe, which are the films "Dr. No" through "The Man with the Golden Gun"
```

```python
bond.iloc[:8] # returns the first 8 rows of the dataframe, which are the films "Dr. No" through "Live and Let Die"
```

```python
bond.iloc[10:] # returns the 11th to the last rows of the dataframe, which are the films "The Spy Who Loved Me" through "Skyfall"
```

```python
bond.iloc[8, 2] # returns the value in the 9th row and 3rd column of the dataframe, which is the budget of the film "The Spy Who Loved Me"
```

```python
bond.iloc[8:12, 2:4] # returns the values in the 9th to 12th rows and 3rd to 4th columns of the dataframe, which are the budgets and box office revenues of the films "The Spy Who Loved Me" through "Skyfall"

```

```python
bond.iloc[3:9, [2, 3, 6]] # returns the values in the 4th to 9th rows and 3rd, 4th, and 7th columns of the dataframe, which are the budgets, box office revenues, and Rotten Tomatoes scores of the films "Dr. No" through "Live and Let Die"
```

```python
bond.iloc[3:9, 3:7] # returns the values in the 4th to 9th rows and 4th to 6th columns of the dataframe, which are the box office revenues, Rotten Tomatoes scores, and Metacritic scores of the films "Dr. No" through "Live and Let Die"
```

```python
bond.iloc[:10, :3] # returns the values in the first 10 rows and first 3 columns of the dataframe, which are the films "Dr. No" through "The Spy Who Loved Me" and their release years and budgets  
```

```python
bond.iloc[20:, :3] # returns the values in the 21st to the last rows and first 3 columns of the dataframe, which are the films "Casino Royale" through "Skyfall" and their release years and budgets
```

## Retrieve Rows by Index Label with loc Accessor
    * the loc accessor retrieves one or more rows by index label.
    * provide a pair of square brackets after the accessor.
    * Note that it is important to sort the index for labeled index so that loc[] will work properly for retrieving a range of indices.

```python
bond.head()
```

```python
bond.set_index('Film', inplace=True)
bond.sort_index(inplace=True) # sorts the dataframe by the index (film titles) in alphabetical order
bond
```

```python
bond.loc["Goldfinger"] # returns the row of the dataframe corresponding to the film "Goldfinger"
```

```python
bond.loc["Casino Royale"] # returns the row of the dataframe corresponding to the film "Casino Royale"
```

```python
bond.loc[["Goldfinger", "Casino Royale"]] # returns the rows of the dataframe corresponding to the films "Goldfinger" and "Casino Royale"
```

```python
bond["Diamond Are Forever":"The Spy Who Loved Me"] # returns the rows of the dataframe corresponding to the films "Diamond Are Forever" through "The Spy Who Loved Me"  
```

```python
bond.loc[:"On Her Majesty's Secret Service"] # returns the rows of the dataframe corresponding to the films from the beginning up to and including "On Her Majesty's Secret Service"    
```

```python
bond.loc["Dr. No":] # returns the rows of the dataframe corresponding to the films from "Dr. No" to the end of the dataframe, which are all the films in the dataframe since "Dr. No" is the first film in alphabetical order
```

```python
bond["Casino Royale":"Skyfall":] 
# returns the rows of the dataframe corresponding to the films "Casino Royale" through "Skyfall", which are all the films in the dataframe 
# since "Casino Royale" is the first film in alphabetical order and "Skyfall" is the last film in alphabetical order
```

```python
bond[:"Casino Royale"] # returns the rows of the dataframe corresponding to the films from the beginning up to and including "Casino Royale"
```

## Overwrite Value in a DataFrame

        * bond.loc["Actor"] returns the row(s) where the index label is "Actor".

```python
bond
```

```python
bond.loc["Diamonds Are Forever", "Actor"] = "Sir Sean Connery"
# updates the value in the "Actor" column for the row corresponding to the film "Diamonds Are Forever" to "Sir Sean Connery"
```

## Overwrite Multiple Values in a DataFrame
    * the replace method replaces all occurances of a Series value with another value (similar to Find & Replace).
    * to overwrite multiple values in a DataFrame, remember to use an accessor on the DataFrame itself.
    * accessors like loc and iloc can accept Boolean Series. Use them to target the values to overwrite.

```python
bond
```

```python
bond["Actor"] = bond["Actor"].replace("Sean Connery", "Sir Sean Connery") # replaces all occurrences of "Sean Connery" with "Sir Sean Connery" in the "Actor" column of the dataframe 
```

```python
bond
```

```python
bond[bond["Actor"] == "Sir Sean Connery"].loc[:, "Actor"] = "Sean Connery" 
# this will not work as expected because it creates a copy of the dataframe and modifies the copy instead of the original dataframe. 
# To modify the original dataframe, you can use the .loc indexer to select the rows where the "Actor" column is equal to "Sir Sean Connery" 
# and then assign the value "Sean Connery" to those rows in the "Actor" column.
```

```python
bond
```

```python
# this will work as expected because it uses the .loc indexer to select the rows where the "Actor" column is equal to "Sir Sean Connery" 
# and then assigns the value "Sean Connery" to those rows in the "Actor" column of the original dataframe.
bond.loc[bond["Actor"] == "Sir Sean Connery", "Actor"] = "Sean Connery"
bond
```

## Rename Index Labels or Columns in a DataFrame
    * the rename method accepts a dictionary for either its columns or index parameters.
    * the dictionary keys represent the existing names and the values represent the new names.
    * we can replace all columns by overwriting the DataFrame's columns attribute.

```python
bond
```

```python
bond.rename(columns={"Year": "Year of Release", "Box Office": "Revenue"}, inplace=False) 
# renames the "Year" column to "Year of Release" in the dataframe and the "Box Office" column to "Revenue" in the dataframe, 
# but does not modify the original dataframe because inplace=False is specified. To modify the original dataframe, you can set 
# inplace=True or assign the result of the rename operation back to the original dataframe like this:
# bond = bond.rename(columns={"Year": "Year of Release", "Box Office": "Revenue"})
```

```python
swaps = {
    "Dr. No": "Dr. No",
    "Goldeneye": "Goldeneye",
    "The World Is Not Enough": "Best Bond Film Ever",
}
bond.rename(index=swaps, inplace=False) 
# renames the index labels "Dr. No" to "Dr. No", "Goldeneye" to "Goldeneye", and "The World Is Not Enough" to "Best Bond Film Ever" in the dataframe, 
# but does not modify the original dataframe because inplace=False is specified. To modify the original dataframe, you can set inplace=True or assign 
# the result of the rename operation back to the original dataframe like this:          
# bond = bond.rename(index=swaps)

# Note that bond.columns[3] = "Revenue" will not work because the columns attribute of a dataframe is an Index object, which is immutable and does 
# not support item assignment.
# To change the column name, you can use the rename method or assign a new list of column names to the columns attribute of the dataframe like this:
# bond.columns = ["Year", "Actor", "Budget", "Revenue", "Rotten Tomatoes Score", "Metacritic Score", "Salary"]
```

## Delete Rows or Columns from a DataFrame
    * the drop method deletes one or more rows/columns from a DataFrame.
    * pass the index or columns parameters a list of the column names to remove.
    * the pop method and removes and returns a single Series (it mutates the DataFrame in the process).
    * python's del keyword also removes a single Series.

```python
bond = pd.read_csv('data/jamesbond.csv', index_col="Film").sort_index()
bond # reads the CSV file 'jamesbond.csv' into a new dataframe called bond, which will overwrite the existing bond dataframe with the original data from the CSV file
```

```python
bond.drop(columns=["Year", "Budget"], inplace=False) 
# drops the "Year" and "Budget" columns from the dataframe and modifies the original dataframe because inplace=True is specified. 

```

```python
bond.drop(index=["Dr. No", "Casino Royale"], inplace=False)
# drops the rows with index labels "Dr. No" and "Casino Royale" from the dataframe and modifies the original dataframe
#  because inplace=True is specified. To modify the original dataframe, you can set inplace=True or assign the result 
# of the drop operation back to the original dataframe like this:
# bond = bond.drop(index=["Dr. No", "Casino Royale"])
```

```python
bond.drop(index=["Dr. No", "Casino Royale"], columns=["Year", "Budget"], inplace=False)
```

```python
bond.pop("Year") 
# removes the "Year" column from the dataframe and returns it as a Series. 
# The original dataframe is modified and the "Year" column is no longer present in the dataframe.
```

```python
bond.head()
```

```python
del bond["Budget"] # removes the "Budget" column from the dataframe and modifies the original dataframe.
```

## Create Random Sample with the sample Method
    * the sample method returns a specified one or more random rows from the DataFrame.
    * customize the axis parameter to extract random columns.

```python
bond = pd.read_csv('data/jamesbond.csv', index_col="Film").sort_index()
bond
```

```python
bond.sample(n=5) 
# returns a random sample of 5 rows from the dataframe, which can be useful for quickly inspecting a subset of the 
# data or for testing purposes. The original dataframe is not modified by this operation. 
```

```python
bond.sample(n=2, axis="rows") 
# returns a random sample of 2 rows from the dataframe, which can be useful for quickly inspecting a subset of the 
# data or for testing purposes. The original dataframe is not modified by this operation.
```

```python
bond.sample(n=2, axis="columns") 
# returns a random sample of 2 columns from the dataframe, which can be useful for quickly inspecting a subset of the 
# data or for testing purposes. The original dataframe is not modified by this operation.
```

## The nsmallest and nlargest Methods
    * the nlargest method returns a specified number of rows with the largest values from a given column.
    * the nsmallest method returns rows with the smallest values from a given column.
    * the nlargest and nsmallest methods are more efficient than sorting the entire DataFrame.
    * only works on numeric values.

```python
bond
```

```python
bond.nlargest(n=4, columns="Box Office") 
```

```python
bond["Box Office"].nlargest(n=4)
# returns the "Box Office" column of the dataframe (a Series), which contains the box office revenues of each film in millions of dollars
```

```python
bond.nsmallest(n=3, columns=["Box Office", "Budget", "Bond Actor Salary"])
# returns the 3 rows of the dataframe with the smallest values in the "Box Office", "Budget", and "Bond Actor Salary" columns, 
# which are the films with the lowest box office revenues, budgets, and actor salaries. The original dataframe is not modified by this operation.  

```

## Filtering with the where Method
    * similar to square brackets or loc, the where method filters the original DataFrame with a Boolean Series.
    * Pandas will populate rows that do not match the criteria with NaN values.
    * leaving in the NaN values can be advantageous for certain merge and visualization operations.

```python
bond.head()
```

```python
is_sean_connery = bond["Actor"] == "Sean Connery"
bond.where(cond=is_sean_connery)
# returns a new dataframe where the rows that satisfy the condition (where the "Actor" column is equal to "Sean Connery") 
# are kept unchanged, and the rows that do not satisfy the condition are replaced with NaN (Not a Number) values. The original 
# dataframe is not modified by this operation.
```

## The apply Method with DataFrames
    * the apply method invokes a function on every column or every row in the DataFrame.
    * pass the uninvoked function as the first argument to the apply method.
    * pass the axis parameter an argument of "columns" to invoke the function on every row.
    * pandas will pass in the row's values as a Series object. We can use accessors like loc and iloc to extract the column's values for that row.

```python
bond.head()
```

```python
bond["Actor"].apply(len)
# applies the len function to each value in the "Actor" column of the dataframe, which returns a new Series where each 
# value is the length of the corresponding actor's name in the "Actor" column. The original dataframe is not modified by this operation.
```

```python
def rank_movie(row):
    year = row.loc["Year"]
    actor = row.loc["Actor"]
    budget = row.loc["Budget"]

    if year >= 1980 and year < 1990 :
        return "Great 80's flick!"
    
    if actor == "Pierce Brosnan":
        return "The best Bond ever!"
    
    if budget > 100:
        return "Expensive movie, fun!"
    
    return "No comment."
    
bond.apply(rank_movie, axis="columns")
# applies the rank_movie function to each row of the dataframe (axis=1), which returns a new Series where each value is the result of 
# applying the rank_movie function to the corresponding row of the dataframe. The original dataframe is not modified by this operation.
```
