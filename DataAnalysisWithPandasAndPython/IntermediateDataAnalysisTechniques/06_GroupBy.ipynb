## The GroupBy Object

```python
import pandas as pd

fortune = pd.read_csv("data/fortune1000.csv", index_col="Rank")
fortune
```

## The groupby Method
    * Grouping is a way to organize/categorize/group the data based on a column's values.
    * the groupby method returns a DataFrameGroupBy object. It resembles a group/collection of DataFrame in a dictionary-like structure.
    * the DataFrameGroupBy object can perform aggregate operations on each group within it.

___
A DataFrameGroupBy object is the invisible "holding pen" that Pandas creates when you use the .groupby() method.

It is important to understand that calling .groupby() does not actually perform a calculation—it just organizes the data into internal "buckets" based on your criteria, waiting for your next instruction.
___
1. The "Split-Apply-Combine" Pattern
    - To understand the GroupBy object, you have to look at the three-step process it facilitates:
        1. Split: Pandas looks at your specified column (e.g., Country) and groups all rows with the same value together.
        2. Apply: You tell Pandas what to do with those groups (e.g., .mean(), .sum(), or .count()).
        3. Combine: Pandas takes the results from each group and glues them back together into a new DataFrame or Series.
___
2. Useful Hidden Features
    - Because the GroupBy object contains all the grouped data, you can inspect it using these methods:
        - .groups: Returns a dictionary where keys are the group names and values are the index labels of the rows in that group.
        - .get_group("Canada"): Allows you to pull out a single "bucket" as its own DataFrame.
        - .size(): Returns a Series showing how many rows are in each group (similar to value_counts).
___

Summary Table of DataFrameGroupBy object methods
|Method         | Returns                                       | Handles NaN ?
|:---            |:---                                          | :---
|`.first()`   | The first valid value in each column per group.   | Yes (skip them)
|`.last()`    | The last valid value in each column per group.    | Yes (skip them)
|`.head(1)`   | The literal first row of the group.               | No (returns the NaN)

```python
# sum of revenue for each sector

sectors = fortune.groupby(by="Sector") # returns a DataFrameGroupBy Object
sectors
```

```python
sectors.groups
```

```python
sectors.size() # display the number of rows for each Sector
```

```python
sectors.first()
```

```python
sectors.last()
```

```python
sectors.head(1)
```

## Retrieve A Group with the get_group Method
    - the get_group method on the DataFrameGroupBy object retrieves a nested DataFrame belonging to a specific group/category.

```python
sectors.get_group(name="Retailing") # returns all the rows belonging to the "Retailing" sector.
```

```python
fortune[fortune["Sector"] == "Retailing"]
```

## Methods on the GroupBy Object
    * use square brackets on the DataFrameGroupBy object to extract a column from the original DataFrame.
    * the resulting SeriesGroupBy object will have aggregation methods available on it.
    * Pandas will perform the calculation on every group within the collection.
    * for example, the 'sum' method will sum together the Revenues for every row by group/category.

```python
sectors["Revenue"] # returns a SeriesGroupBy object
```

```python
sectors["Revenue"].sum() # sum up the Revenue for each sector group.
```

```python
sectors["Employees"].sum()
```

```python
sectors["Profits"].max()
```

```python
sectors[["Revenue","Profits"]].sum()
```

## Grouping by Multiple Columns
    * pass a list of columns to the groupby method to group by pairing of values across columns.
    * target a column to retrieve the SeriesGroupBy object, then perform an aggregation with a method.
    * pandas will return a MultiIndex Series where the levels will be the original groups.

```python
sector_industry = fortune.groupby(by=["Sector","Industry"])
sector_industry
```

```python
sector_industry.size()
```

```python
sector_industry["Revenue"].sum()
```

```python
sector_industry["Employees"].mean()
```

## The agg Method
    * the agg method applies different aggregation methods on different columns.
    * invoke the agg method directly on the DataFrameGroupBy object.
    * pass the method a dictionary where the keys are the columns and the values are the aggregation operations.

```python
sectors.agg({"Revenue":"sum", "Profits":"max"})
```

## Iterating through Groups
    * the DataFrameGroupBy object supports the apply method just like Series and DataFrame.
    * the apply method invokes a function on every nested DataFrame in the DataFrameGroupBy object.
    * it captures the return values of the functions and collects them in a new DataFrame, the return value.

```python
# find the two companies in each sector with the most employees

def top_two_companies_by_employee_count(sector):
    return sector.nlargest(2, "Employees")

sectors.apply(top_two_companies_by_employee_count)
```
