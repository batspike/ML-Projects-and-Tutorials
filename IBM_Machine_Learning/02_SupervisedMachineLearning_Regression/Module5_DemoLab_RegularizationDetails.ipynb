### Additional Notes on Data Standardization and Scaling

When we perform `StandardScaler().fit_transform(X)`, it only scale the data X. It scale it such that the result will end up with a mean of 0, and standard deviation of 1. This does not affect its overall shape. If the original data is skewed, the scaled data will still be skewed. If we want to modify the shape of the data to a standard normal bell shape, a separate transformation process (typically using Log or Box-Cox Transform) needs to be applied.

* **StandardScaler** is strictly a "resizing" tool. It makes the data "standard" (Mean 0, Std 1), but it does not change the shape. If our data was a skewed, lopsided mountain, it will remain a skewed, lopsided mountain—just centered at zero and with a height of 1.

* **Transformation** is the "reshaping" tool. If we want to force our data to look like a bell-shaped (Normal) distribution, that is a separate, prior step.

|Tool|Action|What it changes|Goal|
|---|---|---|---|
|StandardScaler|Scaling|The units/magnitude.|"Keeps features on the same ""playing field"" so no single feature dominates the regularization math."|
|Log/Box-Cox Transform|Transformation|The shape/distribution.|Fixes skewness so the data looks like a bell curve (Normal Distribution).|

<BR>

**Why would you want a "Bell Shape"?**

Many machine learning models—especially Linear Regression, Ridge, and Lasso—make a mathematical assumption called Normality. They assume that our features and the target variable share a linear, symmetrical relationship.

If our data is highly skewed (like House Prices, where most houses are cheap and a few are multi-million dollar mansions), the model will get confused. It will pay too much attention to the expensive outliers because the skewness makes them "loud."

By applying a transformation (like np.log1p or PowerTransformer), we "squash" the long tail of the distribution, pulling those outliers closer to the rest of the pack, creating that nice, clean bell shape.

<BR>

**The "Architect's" Proper Workflow**

In a professional pipeline, you almost always do these in a specific order:

1. **Transformation (The Reshaper)**: We apply np.log1p or PowerTransformer to fix the skewness. Now our data has the right shape.

2. **Scaling (The Resizer)**: We apply StandardScaler. Now our data has the right size/units.

3. **Model Training**: We feed this clean, transformed, and scaled data into our Lasso/Ridge model.

It is important that the Reshaper Transformation phase must happen before Scaling (the Resizer). This is because during Scaling the feature mean and standard deviation values are scaled into 0 and 1 respectively. Applying Reshaper Transformation which usually perform `np.log1p()` on the data will results in errors. To avoid this, always follow the sequence of operations that respects the mathematical requirements of the functions we are using.

|Step|Operation|Purpose|
|---|---|---|
|1. Transformation|Log/Box-Cox|"Fix skewness while data is still in its ""natural"" (usually positive) state."|
|2. Scaling|StandardScaler|Level the playing field for the machine learning model.|

### **LinearRegression Model Coefficients with and without scaling**


#### Without Scaling

````python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

y_col = "MEDV"

X = boston_data.drop(y_col, axis=1)
y = boston_data[y_col]

lr.fit(X, y)
print(lr.coef_) # min = -18
````

### **Interpretation of the above metrics**

The $R^2$ value tells us the **accuracy** of the model (the higher the $R^2$, the higher its accuracy), and the coefficients tell us about the **Model Complexity** and **Reliability**. 

Based on the above numbers, Lasso is the clear winner. Here is how those two specific metrics tell the story of why Lasso is better than the "Standard" Linear Regression in this scenario.

1. Number of Non-zero Coefficients (Feature Selection)

    * Linear Regression (104 features): It is using every single piece of data you gave it. It is "listening" to every bit of noise and every redundant variable.

    * Lasso (90 features): Lasso has "muted" 14 features by forcing their coefficients to zero.

    * What this tells us: Lasso has performed Automatic Feature Selection. It essentially told us, "Hey, 14 of these columns are actually distracting us from the truth." A model that achieves a higher R2 with fewer variables is always superior because it is more efficient and easier to explain.

2. Sum of the Magnitude of Coefficients (The "Stable" Model)

    * Linear Regression (~1183): The weights are very high.

    * Lasso (~435): The weights are significantly smaller (nearly 1/3 the size).

    * What this tells us: High coefficients in a Linear Regression model are often a "red flag" for overfitting. When a model has massive coefficients, it means the prediction is extremely sensitive to tiny changes in the input data (high variance).

    * The "Intensity" Factor: By shrinking the magnitude to 435, Lasso has "calmed the model down." It is relying on more reasonable, stable relationships rather than extreme mathematical swings. This Sum of Magnitude of Coefficients also indicates how sensitive the model output is to its features.

# Example: Does it matter when you scale?


#### Scale then Split

````python
X_train, X_test, y_train, y_test = train_test_split(X_ss, y, test_size=0.3, random_state=72018) # splitting on the Scaled data

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

print("r2 score for Linear Regression (Scaled then Split):", r2_score(y_test,y_pred)) # Calculating r2 score
````

#### Split then Scale

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=72018) # splitting on the Un-scaled data

s = StandardScaler()
lr_s = LinearRegression()
X_train_s = s.fit_transform(X_train)
lr_s.fit(X_train_s, y_train)

X_test_s = s.transform(X_test)
y_pred_s = lr_s.predict(X_test_s)

print("r2 score for Linear Regression (Splitted then Scale):", r2_score(y_test, y_pred_s)) # Calculating r2 score
````

**Conclusion:** It doesn't seem to matter whether we scale before or afterwards, in terms of the raw predictions, for Linear Regression. However, it matters for other algorithms. Plus, as we'll see later, we can make scaling part of a `Pipeline`.

**Note**: It is **NOT** advisable to perform Scaling first before Splitting regardless of model type. Scaling data before Splitting could potentially leads to **Data Leakage**. When we run StandardScaler().fit() on our entire dataset before splitting, the scaler looks at every single row to calculate the mean and standard deviation.

* **The Leak**: Information from your future "Test Set" (which the model is never supposed to see) is now "leaked" into the "Training Set" through those mean and scaling factors.

* **The Consequence**: The model's performance metrics (R2, RMSE) will be artificially inflated. It's like a student getting a peek at the answer key before taking the exam—they look smarter than they actually are.

<BR>

**The Only Correct Workflow**

To be a professional "Architect," we must always follow this sequence:

1. Split the data into Training and Test sets.

2. Fit the Scaler on the Training Set only.

3. Transform both the Training Set and the Test Set using those training parameters.

<BR>

**The Golden Rule:** The model must live in the "universe" created by the training data. The Test set is just a visitor to that universe.
