```python
# Applications with text embeddings
# We will now revisit using language models to calculate the similarities between sequences of text. Below is a corpus (a collection of texts) 
# that we will be embedding into a text vector space. This corpus is a selection of science trivia facts compiled by the BBC.

corpus = ["A cloud weighs around a million tonnes.",
          "Giraffes are 30 times more likely to get hit by lightning than people.",
          "Identical twins don't have the same fingerprints.",
          "Earth's rotation is changing speed.",
          "All the world's bacteria stacked on top of each other would stretch for 10 billion light-years.",
          "You can yo-you in space.",
          "Fish form orderly queues in emergencies.",
          "Most ginger cats are male.",
          "There's a planet made mostly from diamond.",
          "The biggest butterfly in the world has a 31 cm wingspan.",
          "The moon is shrinking.",
          "Flamingoes aren't born pink.",
          "Someone left a family photo on the Moon.",
          "Your brain burns 400-500 calories a day.",
          "You can't fold a piece of A4 paper more than 8 times.",
          "You can smell ants."
]


# Create an array representing the embeddings of each sentence in the corpora. You will need to instantiate a model using the sentence-transformers 
# library. As in the example from the video, use the all-MiniLM-L6-v2 model to encode the corpus. 
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(corpus) 
print(f"\n\nembedings.shape : {embeddings.shape}")
print(f"embedings :\n {embeddings}")

```

```python
# Now, create a target sentence that we want to find "neighbors" of in the embedding vector space. A good example might be "Give me a fact about an animal." 
target = "Give me a fact about an animal."
target_embedding = model.encode(target)

# Now we can calculate the dot product for each of the sentences with the target embedding.
dot_products = np.dot(embeddings, target_embedding)
print(f"\ndot_products :\n {dot_products}" )

# Finally, we can return the sentence with the highest dot product with the target.
most_similar_sentence = corpus[np.argmax(dot_products)]
print(f"\nMost similar sentence :\n {most_similar_sentence}")

```

```python
# corpus sentences ranking in descending order of similarity

dot_products_desc = np.sort(dot_products)[::-1]
for n in dot_products_desc:
    idx_similar_sentence = np.where(dot_products == n)[0]
    print(f"{n} \t-> \t{corpus[idx_similar_sentence[0]]}")

```
