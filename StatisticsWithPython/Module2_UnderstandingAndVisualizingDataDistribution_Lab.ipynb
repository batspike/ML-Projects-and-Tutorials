```python
# Basic probability review

# Review the code notebook from the video lecture about basic probability. Then, write a function named coin_flip() that simulates 
# flipping a fair coin n times. The function should take as an argument n, an integer number of coin flips, and return an array of 
# values representing the outcomes (where 1.0 is heads and 0.0 is tails.)


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import random
from scipy.stats import bernoulli


def coin_flip(n, p=0.5):
    return bernoulli.rvs(p=0.5, size=n)

```

```python
# Use your function to simulate 10 coin flips. Then calculate what percent of the coin flips are heads. Print out the result.

draws = coin_flip(n=10)

df = pd.DataFrame(draws, columns=["Outcome"])
df['Outcome'].value_counts(normalize=True)
```

```python
# Create a new function named loaded_coin_flip(), which takes as arguments:
    # , the number of coin flips to simulate, and
    # , the probability of getting heads

def loaded_coin_flip(n, p=0.5):
    draws = bernoulli.rvs(p=p, size=n)
    df = pd.DataFrame(draws, columns=['Outcome'])
    df['Outcome'] = df['Outcome'].map({1: 'Heads', 0: 'Tails'})
    return df['Outcome'].value_counts(normalize=True)
```

```python
loaded_coin_flip(n=100, p=0.6)
```

```python
def sample_poisson(lam=1, n=1000):
    data = np.random.poisson(lam=lam, size=n)
    return pd.DataFrame(data, columns=["outcome"])
```

```python
df = sample_poisson()
```

```python
def simulate_one_poll(p, n_sample, seed = 1):
    responses = stats.bernoulli.rvs(p = p, size = n_sample, random_state = seed)
    responses = pd.DataFrame(responses, columns = ['response'])
    return responses

results = simulate_one_poll(0.5, 500)
```

```python
results['response'].describe()
```

```python
poll_a = simulate_one_poll(0.5, 500)
poll_b = simulate_one_poll(0.7, 500)
```
