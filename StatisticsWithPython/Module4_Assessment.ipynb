```python
import pandas as pd
import numpy as np
import seaborn as sns

rng = np.random.default_rng(12) # seed for reproducibility. do not remove.
penguins = sns.load_dataset("penguins")
```

## Challenge 01
### 1.1 Filter penguins by species [1 pt]
Create a new DataFrame, `adelie` with only penguins of the species "Adelie" who live on Biscoe or Dream islands. Then, reduce the DataFrame to two columns, `island` and `body_mass_g`.

```python
adelie = penguins[penguins['species'] == 'Adelie']
adelie = adelie[adelie['island'].isin({'Biscoe','Dream'})][['island','body_mass_g']]
```

### 1.2  Calculate the average body mass of each species [1 pt]
Calculate the average body mass of the penguins on each island and then calculate the difference using the formula $ \bar{m}_{Biscoe} -\bar{m}_{Dream}$, where $ \bar{m}_{Biscoe}$ and $ \bar{m}_{Dream}$ are the average body mass of penguins on Biscoe and Dream islands, respectively. Save the difference in a variable called `observed_mass_diff`.

*Terminology note:* The quantity $\bar{m}_{Biscoe} -\bar{m}_{Dream}$ is called a "test statistic".

```python
mu_Biscoe = adelie[adelie['island'] == 'Biscoe']['body_mass_g'].mean()
mu_Dream = adelie[adelie['island'] == 'Dream']['body_mass_g'].mean()

observed_mass_diff = mu_Biscoe - mu_Dream
```

## Challenge 02
Is the observed difference in Adelie penguin body mass between Biscoe and Dream islands statistically significant? In other words, let's conduct a hypothesis test where:

- **Null Hypothesis ($H_0$):** There is no difference in body mass between Adelie penguins on Biscoe and Dream islands.
- **Alternative Hypothesis ($H_A$):** There is a difference in body mass between Adelie penguins on Biscoe and Dream islands.


We will use a non-parametric permutation test. As a first step, create helper functions to shuffle data and calculate the test statistic for a sample.

### 2.1 Create a function to shuffle island labels [1 pt]
Create a function called `shuffle_labels` that shuffles the `island` values for a DataFrame. The function should take a DataFrame as input and return a new DataFrame with the same data but with the island labels randomly shuffled.

```python
def shuffle_labels(df):
    """
    Shuffles the 'island' column of the input DataFrame.
    Returns a copy of the DataFrame with shuffled labels.
    """
    # Create a copy to avoid modifying the original dataframe
    shuffled_df = df.copy()
    
    # Extract the island column, shuffle it, and reset the index
    # so it can be reassigned to the dataframe
    shuffled_df['island'] = df['island'].sample(frac=1).values
    
    return shuffled_df
```

### 2.2 Create a function to calculate the difference in body mass means [1 pt]
Next, define a function called `calculate_mass_diff` that calculates the test statistic--the difference in average body mass between Biscoe and Dream Island penguin--for a given DataFrame. The function should take a DataFrame as input and return the difference in average body mass between Biscoe and Dream islands, formatted as an `np.float`.

```python
def calculate_mass_diff(df):
    """
    Calculates the difference in average body mass (Biscoe - Dream) 
    for the penguins in the provided DataFrame.
    """
    # Calculate means for each island group
    mu_biscoe = df[df['island'] == 'Biscoe']['body_mass_g'].mean()
    mu_dream = df[df['island'] == 'Dream']['body_mass_g'].mean()
    
    # Calculate the difference (Biscoe - Dream)
    diff = mu_biscoe - mu_dream
    
    # Return as np.float
    return np.float64(diff)
```

## Challenge 03
### 3.1 Simulate the null distribution [2 pts]
Using the helper functions we just created, simulate the null distribution of our test statistic using 10,000 permutations. Save the results of the simulation in a variable called `sample_diffs`. This variable should be a DataFrame with one column, `statistic`, that contains the calculated test statistic for each permutation.

```python
# Perform 10,000 permutations
n_permutations = 10000
stats = []

for i in range(n_permutations):
    # Shuffle the labels
    shuffled = shuffle_labels(adelie)
    # Calculate the test statistic for the shuffled data
    stat = calculate_mass_diff(shuffled)
    stats.append(stat)

# Save the results into a DataFrame
sample_diffs = pd.DataFrame(stats, columns=['statistic'])
```

## Challenge 04
### 4.1 Calculate the p-value of the observed difference [1 pt]
Calculate the two-sided p-value of the observed difference in body mass between Biscoe and Dream islands. Save the p-value in a variable called `p_value`.

```python
p_value = (sample_diffs['statistic'].abs() >= abs(observed_mass_diff)).mean()
```

### 4.2 Interpret significance [1 pt]
Given the standard significance threshold of $ \alpha = 0.05$ and the p-value you calculated, can you reject the null hypothesis that there is no difference in body mass between Adelie penguins on Biscoe and Dream islands? Set the value of the variable `reject_null` to a Boolean representing your answer (i.e., `True` if you would reject the null hypothesis)

Based on the p-value of approximately 0.8253 and the significance threshold of α=0.05, we fail to reject the null hypothesis.

Since the p-value is much greater than 0.05, the observed difference in body mass between the two islands is not statistically significant and is likely due to random sampling variation.
